```python
from pymongo import MongoClient
import pandas as pd
import numpy as np
import ast
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
healthy = pd.read_csv('../data/twitter_mongo/healthy_final.csv',index_col=0)
unhealthy = pd.read_csv('../data/twitter_mongo/unhealthy_final.csv',index_col=0)
grocery = pd.read_csv('../data/twitter_mongo/grocery_stores_final.csv',index_col=0)
ff_stores = pd.read_csv('../data/twitter_mongo/ff_stores_final.csv',index_col=0)
```

```python
healthy.shape[0]+unhealthy.shape[0]+grocery.shape[0]+ff_stores.shape[0]
```

```python
healthy = healthy[(healthy['long']>-131)&(healthy['long']<-68)&(healthy['lat']>20)&(healthy['lat']<64)]
unhealthy = unhealthy[(unhealthy['long']>-131)&(unhealthy['long']<-68)&(unhealthy['lat']>20)&(unhealthy['lat']<64)]
grocery = grocery[(grocery['long']>-131)&(grocery['long']<-68)&(grocery['lat']>20)&(grocery['lat']<64)]
ff_stores = ff_stores[(ff_stores['long']>-131)&(ff_stores['long']<-68)&(ff_stores['lat']>20)&(ff_stores['lat']<64)]
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,7))
ax.scatter(healthy['long'],healthy['lat'],c='b',alpha=.4,label = "Healthy words")
ax.scatter(grocery['long'],grocery['lat'],c='y',alpha=.4,label = "Grocery stores")
ax.set_title("Positive Health Tweets")
ax.set_xlabel("Latitude")
ax.set_ylabel("Longitude")
ax.legend()
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,7))
ax.scatter(unhealthy['long'],unhealthy['lat'],c='purple',alpha=.4, label = "Unheatlhy words")
ax.scatter(ff_stores['long'],ff_stores['lat'],c='orange',alpha=.4,label="Fast Food stores")
ax.set_title("Negative/Unhealthy Tweets")
ax.set_xlabel("Latitude")
ax.set_ylabel("Longitude")
ax.legend()
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,7))
ax.scatter(healthy['long'],healthy['lat'],c='b',alpha=.4,label="Healthy")
ax.scatter(unhealthy['long'],unhealthy['lat'],c='g',alpha=.4,label="Unhealthy")
ax.scatter(grocery['long'],grocery['lat'],c='y',alpha=.4,label="Grocery")
ax.scatter(ff_stores['long'],ff_stores['lat'],c='r',alpha=.4,label="Fasst Food")
ax.set_xlabel("Latitude")
ax.set_ylabel("Longitude")
ax.legend()
ax.set_title("All Tweets")
```
