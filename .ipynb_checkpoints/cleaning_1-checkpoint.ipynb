```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score, f1_score, accuracy_score,recall_score,precision_score
import warnings
warnings.filterwarnings('ignore')
from sklearn.naive_bayes import BernoulliNB
from sklearn.feature_selection import SelectPercentile, chi2
from sklearn import svm
```

```python
food_desert = pd.ExcelFile('data/DataDownload2015.xlsx')
desert_data = food_desert.parse(2)
desert_data.set_index('CensusTract', inplace=True)
```

```python
behav = pd.read_csv('data/500_Cities__Census_Tract-level_Data__GIS_Friendly_Format___2018_release.csv',)
behav.set_index('TractFIPS', inplace=True)
```

```python
merge1 = behav.merge(desert_data, how='left', left_index=True, right_index=True)
```

```python
merge1.head()
```

```python
merge1['LILATracts_halfAnd10'].fillna(0,inplace=True)
```

```python
y=merge1.pop('LILATracts_halfAnd10')
```

```python
X=merge1
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.head()
```

```python
def clean(df):
    df.fillna(df.mean(),inplace=True)
    data1=df[['TEETHLOST_CrudePrev','STROKE_CrudePrev','PHLTH_CrudePrev','OBESITY_CrudePrev','MHLTH_CrudePrev','LPA_CrudePrev','KIDNEY_CrudePrev','DENTAL_CrudePrev','DIABETES_CrudePrev','CSMOKING_CrudePrev','COREW_CrudePrev','COPD_CrudePrev','CHOLSCREEN_CrudePrev','CASTHMA_CrudePrev','ACCESS2_CrudePrev','PovertyRate','TractSNAP']]
    return data1
```

```python
df=clean(X_train)
```

```python
df.describe()
```

```python
df_test = clean(X_test)
```

```python
df_test.head()
```

## REMOVE AFTER THIS

```python
model_log = LogisticRegression(penalty='l1')
model_log.fit(df,y_train)
```

```python
print("TRAIN")
print("ROC_AUC Logistic: ", cross_val_score(LogisticRegression(),df,y_train,cv=5,scoring='roc_auc').mean())
print("Accuracy Logistic: ", cross_val_score(LogisticRegression(),df,y_train,cv=5).mean())
print("f1 Logistic: ", cross_val_score(LogisticRegression(),df,y_train,cv=5,scoring='f1').mean())
```

```python
y_pred = model_log.predict(df_test)
```

```python
print("TEST")
print("ROC_AUC Logistic: ", roc_auc_score(y_test,y_pred))
print("Accuracy Logistic: ", accuracy_score(y_test,y_pred))
print("f1 Logistic: ", f1_score(y_test,y_pred))
```
