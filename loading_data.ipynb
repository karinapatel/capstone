### Consumer expenditure data

```python
consumer = pd.read_csv('data/food_consumer.csv')
```

```python
consumer.columns=['Item', 'midwest_tot', '< 5000',
       '[5000,9999]', '[10000,14999]',
       '[15000,19999]', '[20000,29999]',
       '[30000,39999]', '[40000,49999]',
       '[50000,69999]', '>= 70,000']
```

```python
consumer['Item']=['total_spend','food_home','food_out']
```

```python
consumer.set_index("Item", inplace=True)
```

```python
def convert_int(df):
    return df.apply(lambda x: int(x.replace(",","")))
```

```python
consumer = consumer.apply(convert_int)
```

```python
consumer.info()
```

```python
consumer
```

```python
consumer.loc['food_home'] = consumer.loc['food_home']/consumer.loc['total_spend']
```

```python
consumer.loc['food_out'] = consumer.loc['food_out']/consumer.loc['total_spend']
```

```python
consumer.loc['food_out'].values + consumer.loc['food_home'].values
```
